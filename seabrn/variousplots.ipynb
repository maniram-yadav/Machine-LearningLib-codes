```python
import pandas as pd
import seaborn as sn
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/home/maniram/data/iris.csv")
df.head()
df.columns = ['sepal.length','sepal.width','petal.length','petal.width','class']
```

```python
df.head()
```

```python
%matplotlib inline
sn.jointplot(x='sepal.length',y='sepal.width',data=df,size=10)
sn.jointplot(x='petal.length',y='petal.width',data=df,size=10)
```

```python
sn.FacetGrid(df,hue='class',size=8)\
    .map(plt.scatter,'petal.length','petal.width')\
    .add_legend()
```

```python
sn.boxplot(x='class',y='petal.length',data=df)
plt.show()
sn.boxplot(y='class',x='petal.length',data=df)
```

```python
ax=sn.boxplot(x='class',y='petal.length',data=df)
ax=sn.stripplot(x='class',y='petal.length',data=df,jitter=True,edgecolor='gray')
```

```python
sn.violinplot(x='class',y='petal.length',data=df,)
```

```python
sn.pairplot(df)
```
