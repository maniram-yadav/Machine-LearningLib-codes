```python
ebola = pd.read_csv('../datas/ebola_country_timeseries.csv')
```

```python
ebola.head()
```

```python
eboladata=pd.melt(ebola,
       id_vars=['Date','Day'],
       var_name='cd_country',
       value_name='count').head()
```

```python
ebola_split=eboladata['cd_country'].str.split('_')
```

```python
ebola_split
```

```python
cases=ebola_split.str.get(0)
cases
```

```python
country=ebola_split.str.get(1)
country
```

```python
eboladata['country']=country
eboladata['status']=cases
```

```python
eboladata
```

```python
eboladata=eboladata.drop('cd_country',axis=1)
```

```python
eboladata
```

Above procedure in single step

```python
eboladata=pd.melt(ebola,id_vars=['Date','Day'],var_name='country',value_name='count')
```

```python
ebola_split=eboladata['country'].str.split('_',expand=True)
eboladata=eboladata.drop('country',axis=1)
ebola_split.columns=['status','country']
eboladata=pd.concat([eboladata,ebola_split],axis=1).head()
eboladata=eboladata.fillna(method='ffill')
eboladata
```

#  Common Columns

```python
weather = pd.read_csv('../datas/weather.csv')
weather.head()
```

```python
weatherdata=pd.melt(weather,
                               id_vars=['id','year','month','element'],
                               var_name='day',
                               value_name='temp')
weather_arrange=weatherdata.pivot_table(
    index=['id', 'year', 'month', 'day'],
    columns='element',
    values='temp')
weather_arrange
```

```python
weather_arrange.reset_index()
```
